```python
def jsonToDF(name):
    """Read a list of sentences from the JSON file, store them in a dataframe"""
    
    with open(f"{name}.json") as fin:
        textList = json.load(fin)

    # create a name for each document, based on its category
    indexNames = [f"{name}" for i in range(len(textList))]

    # create the dataframe, it will have one column and one index
    df = pd.DataFrame(data=textList, index=indexNames)
    df.columns = ['document']
    return df
```

```python
def getHashtags(text):
  """Takes a string phrase and returns hashtags if present"""
  if isinstance(text, str):
    return re.findall(r"#(\S+)", text.lower())
  return []
```

```python
#extracts the hashtags for each post
allDocs['hashtags'] = allDocs['document'].apply(getHashtags)
allDocs.head()
allDocs
```

```python
def splitHashtags(sentence):
    """Takes a sentence and splits hashtags if present"""
    if isinstance(sentence, str):
        hashtags = [tag.strip('#') for tag in sentence.split('#') if tag.strip('#')]
        return hashtags
    return []
```

```python
RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)

def strip_emoji(text):
    return RE_EMOJI.sub(r'', text)
```

```python
#toRemove = ['fyp', '#trending', 'trending', '#foryou', 'foryou', '#viral', 'viral', 'foryoupage', 'fy', 'fypage', 'tiktok', 'video', 'forypu', 'fup', 'everyonefyp', 'reaction', 'fypviral']
```

```python
toRemove = []

def remove_words(sentence):
    #if isinstance(sentence, str):
        sent = strip_emoji(sentence)
        hashtags = splitHashtags(sent)
        final = []
        for hash in hashtags:
            #print(type(hash))
            if hash.lower() not in toRemove and hash not in final:
                final.append(hash)
        return ' '.join(final)

allDocs['document'] = allDocs['document'].apply(remove_words)
pd.set_option("display.max_colwidth",1000)
allDocs.head()
```

```python
# create new df for each user

user_documents = {f'user{i}': [] for i in range(1, 11)}
user_hashtags = {f'user{i}': [] for i in range(1, 11)}

for i, row in enumerate(allDocs.itertuples(), start=1):
    user_id = f'user{i % 10 + 1}'  # Assuming there are 10 users and user IDs start from user1
    document = row.document
    hashtags = row.hashtags
    
    user_documents[user_id].append(document)
    user_hashtags[user_id].extend(hashtags)

for user in user_documents:
    user_documents[user] = ' '.join(user_documents[user])
    user_hashtags[user] = ' '.join(user_hashtags[user])

user_documents_df = pd.DataFrame.from_dict(user_documents, orient='index', columns=['documents'])
user_hashtags_df = pd.DataFrame.from_dict(user_hashtags, orient='index', columns=['hashtags'])

user_documents_df['hashtags'] = user_hashtags_df['hashtags']
user_documents_df
```

```python
# exclude user1, 6, 10 since they do not have enough data - content viewing history < 1000

users_to_include = ['user2', 'user3', 'user4', 'user5', 'user7', 'user8', 'user9']

# Filter the DataFrame to include only the specified users
user_documents_df = user_documents_df[user_documents_df.index.isin(users_to_include)]

user_documents_df
```

### Convert to document-term matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Define parameters
tfidf_params = {
    'strip_accents': 'unicode',
    'stop_words': 'english',
    'lowercase': True,
    'token_pattern': r'\b[a-zA-Z]{3,}\b', # we want only words that contain letters and are 3 or more characters long
    'min_df': 2,
    'max_df': 0.7, 
}

# Initialize TfidfVectorizer with the defined parameters
tfidf_vectorizer = TfidfVectorizer(**tfidf_params)

# Transform our data into the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(user_documents_df['documents'])
tfidf_matrix
```

```python
import pandas as pd

# Assuming tfidf_matrix and tfidf_vectorizer are already defined
# Create tfidf_df without user IDs as index
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Define the user IDs
user_ids = ['user2', 'user3', 'user4', 'user5', 'user7', 'user8', 'user9']

# Set the user IDs as index for tfidf_df
tfidf_df.index = user_ids

tfidf_df
```

```python
# Get the top N terms with highest TF-IDF scores for each user

N = 10

top_terms_per_user = {}
for user, row in tfidf_df.iterrows():
    top_terms = row.sort_values(ascending=False)[:N]
    top_terms_per_user[user] = top_terms

# Plot top terms for each user with a taller plot
plt.figure(figsize=(12, 10))  # Adjust the height to be taller
for i, (user, top_terms) in enumerate(top_terms_per_user.items(), start=1):
    plt.subplot(5, 2, i)
    top_terms.plot(kind='bar')
    plt.title(f'Top {N} TF-IDF Terms for User {user}')
    plt.xlabel('Term')
    plt.ylabel('TF-IDF Score')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
